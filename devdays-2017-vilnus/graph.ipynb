```python
%pylab inline
import gzip, pickle
```

```python
with gzip.open("/storage/social_graph/LargeVis/Linux/largevis.txt.gz") as fin:
    size = int(fin.readline().decode().split(" ")[0])
    graph = zeros((size, 2), float32)
    for line in fin:
        parts = line[:-1].decode().split(" ")
        graph[int(parts[0])] = float(parts[1]), float(parts[2])
```

```python
graph[graph > 100] = 0
graph[graph < -100] = 0
```

```python
rcParams["figure.figsize"] = (16, 16)
axis("off")
hexbin(graph[:, 0], graph[:, 1], gridsize=1024, bins="log", cmap="inferno")
subplots_adjust(bottom=-0.1, top=1.1, left=-0.1, right=1.1)
```
